```python
def send_welcome_email(name):
  print(f"Hi,{name}")
  print("welcome to the team \n")

send_welcome_email("Alice")
send_welcome_email("Bob")
send_welcome_email("Charlie")
```

```python
def addition(num1,num2):
  return num1+num2

res=addition(5,9)
print(res)
```

```python
def rev(txt):
  res=""
  for char in txt:
    res=char+res
  return res

rev("Shaan")
```

```python
def cleaning(text):
  clean=text.strip().upper()
  return clean
"""removes whitespaces and transforms to upper or lower
lstrip removes leading and rstrip trailing"""

raw="  AlICe  "
res=cleaning(raw)
print(res)
```

```python
#keyword argguments

def perform(user,action):
  print(f"{user} did {action}")

perform(action="login",user="Alex")
#postional arguments
def perform1(user,action):
  print(f"{user} did {action}")

perform1("alex","login")
```

```python
#default arguments

def default(user,action="login"):
  print(f"{user} did {action}")

default("Alex")
default("alice","logout") #overrdide default
```

```python
#passing lists and tuples as function arguments
def total_revenue(amounts):
  total=0
  for value in amounts:
    total+=value
  return total

amounts=[10,20,30,40,50]
res=total_revenue(amounts)
print(res)
```

```python
#modifying dict in place
def modify_dict(user):
  if user["lifetime value"]>50000:
    user["Status"]="VIP"

customer={"name":"Alex","lifetime value":60000}
modify_dict(customer)
print(customer)
```
