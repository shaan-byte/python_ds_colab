```python
lists=[1,2,3,4,5]
lists.remove(1) #removed valuse specified in the argument
lists.pop() # remove last indedx element or specify index to remove that index value
lists.insert(2,79) #insert 79 at 2nd index
lists.append(100) #append value at last
print(lists)
print(lists[3])
print(lists[-1]) #access last elemet
```

```python
nums=[12,23,3,5,6,787,87,3]
print(len(nums)) #print lengths of list
#Slicing
print(nums[1:4]) #print elements from 1 to 3
print(nums[:4]) # print the first 4 elements
print(nums[::2]) #print every 2 element starting from 0th index
#Searching
print(nums.count(3)) # number of times 3 appears in the list
print(nums.index(87)) # gives index of the specified element
print(3 in nums) #if 3 is present in nums list
nums[5]=22
print(nums)
```

```python
#Tuple unpacking
tuples=(1,2,3,4,5)
tuples2=(100,200,300,400,500)
a,b,*c=tuples2 #tuple unpacking packs the rest elemets as a list
x,y,z,p,q=tuples #tuple unpacking need one variable for every tuple element
print(x,y,z)
print(tuples)
print(a,b,c)
```

```python
#list to suple and vice versa

nums=[1,2,3,4,5]
tup=(10,20,30,40)
nums_tuple=tuple(nums) #convert to tuple
tup_list=list(tup) #convert to list
print(type(nums_tuple))
print(type(tup_list)) #print type

tuple1=(73,54,76,23) #tuples and list concatenation
tuple2=(22,45)
merge_tuple12=tuple1+tuple2
print(merge_tuple12)
```

```python
#dictionaries

dict1={"user":"Shaan","Sibling":"Saaraa","number":22343545}
print(dict1["user"]) #Access element using key
print(dict1.get("Sibling")) #access
dict1.pop("number") #remove from dict using key
dict1["hobby"]="football" #inserting an element
dict1["user"]="arshan" #upadting using a key
print("user" in dict1) #check if key is present in dict
print(dict1.keys()) #print keys
print(dict1.values()) #print values
print(dict1)
```
