# Part 2 — Regex Basics in Pandas

## What is a regex?

A **regular expression** (regex) is a pattern written in a special mini-language that describes a set of strings. Instead of searching for an exact string like `"phone"`, you describe a shape: "any sequence of 10 digits" or "anything that looks like an email address".

You do not need to memorise regex — you need to be able to read common patterns and know where to look up the rest.

## The building blocks you will actually use

| Pattern | Matches | Example |
|---|---|---|
| `\d` | Any digit (0-9) | `\d\d\d` matches `123` |
| `\d+` | One or more digits | `\d+` matches `9876543210` |
| `\d{10}` | Exactly 10 digits | `\d{10}` matches a 10-digit phone |
| `\s` | Any whitespace | `\s+` matches spaces, tabs |
| `\w` | Any word character (letter, digit, `_`) | `\w+` matches a word |
| `.` | Any single character except newline | `a.b` matches `aXb`, `a1b` |
| `[abc]` | Any one of: a, b, or c | `[aeiou]` matches a vowel |
| `[^abc]` | Any character NOT in: a, b, c | `[^\d]` matches non-digits |
| `^` | Start of string | `^\d` starts with a digit |
| `$` | End of string | `\d$` ends with a digit |
| `+` | One or more | `\d+` = one or more digits |
| `*` | Zero or more | `\d*` = zero or more digits |
| `?` | Zero or one (optional) | `\+?` = optional `+` |
| `()` | Capture group — extract this part | `(\d{10})` captures 10 digits |

````python
# The phone column has numbers in many formats. Our goal: extract EXACTLY 10 digits.
# Strategy: strip everything that is NOT a digit, then take the last 10 digits.

print("Phone numbers before cleaning:")
print(customers["phone"].head(10).tolist())
print()

# Step 1: remove all non-digit characters with .str.replace()
# [^\d] means "any character that is NOT a digit"
digits_only = customers["phone"].str.replace(r"[^\d]", "", regex=True)
print("After removing non-digits:")
print(digits_only.head(10).tolist())
print()

# Step 2: take the last 10 characters (handles country codes like 91 prepended)
customers["phone_clean"] = digits_only.str[-10:]

print("Final cleaned 10-digit phone:")
print(customers[["phone", "phone_clean"]].head(10))
````

````python
# .str.extract() — capture a specific part of a string matching a pattern
# The pattern must have at least one () capture group

# Extract the domain from each email address (everything after @)
# Pattern: @ followed by everything until end of string
domain_pattern = r"@([\w\.]+)$"

customers["email_domain"] = customers["email_clean"].str.extract(domain_pattern)

print("Email domains extracted:")
print(customers[["email_clean", "email_domain"]].head(10))
print()
print("Domain distribution:")
print(customers["email_domain"].value_counts())
````

````python
# .str.match() — True if the ENTIRE string matches the pattern from the start
# .str.contains() — True if the pattern appears ANYWHERE in the string

# Validate that the cleaned phone is exactly 10 digits
valid_phone = customers["phone_clean"].str.match(r"^\d{10}$")

print(f"Valid 10-digit phones: {valid_phone.sum()} out of {len(customers)}")
print()

# Mark invalid phones
customers["phone_valid"] = valid_phone
invalid = customers[~customers["phone_valid"]]
if len(invalid) > 0:
    print("Invalid phone numbers:")
    print(invalid[["name_clean", "phone", "phone_clean"]])
else:
    print("All phones are valid 10-digit numbers.")

print()

# Validate cleaned email — must have format: chars @ chars . chars
email_pattern = r"^[\w\.]+@[\w\.]+\.[a-z]{2,}$"
customers["email_valid"] = customers["email_clean"].str.match(email_pattern)
print(f"Valid emails: {customers['email_valid'].sum()} out of {len(customers)}")
print(customers[~customers["email_valid"]][["name_clean", "email", "email_clean"]])
````

# Part 3 — Cleaning Categorical Data

## What the problem looks like

When you run `.value_counts()` on a categorical column and see this:

```
Gold       6
gold       3
GOLD       2
Silver     3
SILVER     2
silver     2
Sliver     1   <- typo
```

You have 7 distinct values but really only 3 categories. Every analysis — groupby, pivot table, value_counts — will treat these as 7 separate things, not 3. Fixing this is called **normalising** or **standardising** categorical data.

The standard workflow:
1. `.value_counts()` — see all the variants
2. `.str.strip().str.title()` — fix casing and whitespace (handles 80% of cases)
3. `.replace()` — fix remaining typos and non-standard spellings
4. `.value_counts()` again — confirm the fix worked

````python
# Step 1: see the problem clearly

print("membership_tier BEFORE cleaning:")
print(customers["membership_tier"].value_counts())
print()
print(f"Distinct values: {customers['membership_tier'].nunique()}")
print("(Should be 4: Gold, Silver, Platinum, Bronze)")
````

````python
# Step 2: fix casing and whitespace in one expression

customers["tier_clean"] = customers["membership_tier"].str.strip().str.title()

print("After .strip().title():")
print(customers["tier_clean"].value_counts())
print()
print(f"Distinct values: {customers['tier_clean'].nunique()}")
print("Better — but 'Sliver' (typo for Silver) is still there.")
````

````python
# Step 3: fix remaining typos with .replace()

customers["tier_clean"] = customers["tier_clean"].replace({
    "Sliver": "Silver"    # typo fix
})

print("After .replace() for typos:")
print(customers["tier_clean"].value_counts())
print()
print(f"Distinct values: {customers['tier_clean'].nunique()}")
print("4 categories, as expected.")
````

````python
# Converting a cleaned categorical column to Pandas Categorical dtype
# Benefits: memory efficiency, natural ordering, explicit allowed values

tier_order = ["Bronze", "Silver", "Gold", "Platinum"]  # logical order, low to high

customers["tier_cat"] = pd.Categorical(
    customers["tier_clean"],
    categories = tier_order,
    ordered    = True           # Bronze < Silver < Gold < Platinum
)

print("tier_cat dtype:", customers["tier_cat"].dtype)
print()
print("Sorting by tier (uses the logical order, not alphabetical):")
print(customers[["name_clean", "tier_cat"]].sort_values("tier_cat"))
print()
print("Customers with Gold or higher:")
print(customers[customers["tier_cat"] >= "Gold"][["name_clean", "tier_cat"]])
````

# Part 4 — Standardizing Labels

## Beyond simple cleaning

So far we have fixed surface-level issues — casing, spaces, typos. Standardizing labels goes a step further: you are making a deliberate decision about how you want your categories to look in reports, visualisations, and exported files.

Common standardization tasks:
- Renaming categories to business-friendly labels
- Grouping many small categories into broader buckets
- Creating a consistent abbreviation scheme
- Handling unexpected / unknown values safely

````python
# Renaming categories to report-friendly labels
# Use case: internal codes need to become human-readable labels for a dashboard

# Simulate a column of internal department codes
customers["dept_code"] = ["ENG", "SLS", "HR", "FIN", "ENG", "SLS",
                           "HR",  "ENG", "FIN", "SLS", "ENG", "HR",
                           "FIN", "ENG", "SLS", "FIN", "HR",  "ENG", "SLS", "FIN"]

label_map = {
    "ENG": "Engineering",
    "SLS": "Sales",
    "HR":  "Human Resources",
    "FIN": "Finance"
}

customers["department"] = customers["dept_code"].map(label_map)

print("Codes mapped to full department names:")
print(customers[["dept_code", "department"]].head(8))
print()
print(customers["department"].value_counts())
````

````python
# Bucketing: grouping many specific values into broader categories
# Use case: email domains collapsed into provider groups

def classify_domain(domain):
    if pd.isna(domain):
        return "Unknown"
    elif domain in ["gmail.com"]:
        return "Google"
    elif domain in ["yahoo.com"]:
        return "Yahoo"
    elif domain in ["hotmail.com", "outlook.com"]:
        return "Microsoft"
    elif domain.endswith(".in"):
        return "Corporate India"
    else:
        return "Other"

customers["email_provider"] = customers["email_domain"].apply(classify_domain)

print("Email domains bucketed into provider groups:")
print(customers[["email_domain", "email_provider"]].value_counts())
````

````python
# Handling unexpected / unknown values safely
# Real data always brings values you did not plan for
# Strategy: use .isin() to validate, then mark anything outside the valid set

valid_tiers = ["Bronze", "Silver", "Gold", "Platinum"]

# Any value NOT in valid_tiers is flagged as Unknown
# np.where: if condition is True use first value, else second
customers["tier_validated"] = np.where(
    customers["tier_clean"].isin(valid_tiers),
    customers["tier_clean"],
    "Unknown"
)

print("Validated tier (any invalid value -> 'Unknown'):")
print(customers["tier_validated"].value_counts())
print()
print("Checking: all values are from the allowed set + 'Unknown':")
allowed = set(valid_tiers) | {"Unknown"}
print(f"All valid: {set(customers['tier_validated'].unique()).issubset(allowed)}")
````

````python
# Putting it all together: standardize feedback into a Sentiment label
# This combines regex detection with categorical standardization

positive_words = ["great", "excellent", "amazing", "love", "happy",
                  "fantastic", "good", "smooth", "recommended", "satisfied"]
negative_words = ["slow", "bad", "worst", "horrible", "disappointed",
                  "poor", "pathetic", "not acceptable", "avoid"]

positive_pat = "|".join(positive_words)
negative_pat = "|".join(negative_words)

def classify_sentiment(text):
    text_lower = text.lower()
    is_positive = bool(re.search(positive_pat, text_lower))
    is_negative = bool(re.search(negative_pat, text_lower))

    if is_positive and not is_negative:
        return "Positive"
    elif is_negative and not is_positive:
        return "Negative"
    elif is_positive and is_negative:
        return "Mixed"
    else:
        return "Neutral"

customers["sentiment"] = customers["feedback"].apply(classify_sentiment)

print("Sentiment distribution:")
print(customers["sentiment"].value_counts())
print()
print("Sample classifications:")
print(customers[["feedback", "sentiment"]].head(10))
````
