# PART A — CSV: Working with Structured Data

## What is structured data?

Data is **structured** when it fits neatly into rows and columns — like a spreadsheet or a table in a database. Every row represents one record. Every column represents one attribute. The structure is rigid and consistent.

| Title | Author | Year | Genre | Rating |
|---|---|---|---|---|
| The Alchemist | Paulo Coelho | 1988 | Fiction | 4.2 |
| Sapiens | Yuval Noah Harari | 2011 | Non-fiction | 4.4 |
| Atomic Habits | James Clear | 2018 | Self-help | 4.7 |

Every row has the same columns. Every column has the same type of value. That consistency is what makes it "structured".

---

## What is a CSV file?

**CSV** stands for **Comma-Separated Values**. It is the simplest possible format for storing structured data as plain text.

The idea:
- Each row of the table becomes one line of text
- The values in each row are separated by commas
- The first line is usually the **header** — the column names

The table above, as a CSV, looks like this:

```
title,author,year,genre,rating
The Alchemist,Paulo Coelho,1988,Fiction,4.2
Sapiens,Yuval Noah Harari,2011,Non-fiction,4.4
Atomic Habits,James Clear,2018,Self-help,4.7
```

That is it. No special software needed to open it — it is just text.

### Why CSV is everywhere

CSV is the most common data exchange format in the world for tabular data. Reasons:
- Every tool understands it — Excel, Google Sheets, Python, R, databases
- It is human-readable
- It is tiny in size compared to Excel files
- It requires no special parsing library for the basics

As a data scientist, you will work with CSV files constantly.

## A.1 — Reading a CSV File

Python's built-in `csv` module handles CSV files. You do not need to install anything.

There are two main ways to read a CSV:

| Reader | What it gives you | Best for |
|---|---|---|
| `csv.reader` | Each row as a **list** of strings | When column order is fixed and known |
| `csv.DictReader` | Each row as a **dictionary** | Almost always — column names become keys |

````python
# First, let's create our CSV file on disk so we have something to read

csv_content = """book_id,title,author,year,genre,rating,available
B001,The Alchemist,Paulo Coelho,1988,Fiction,4.2,yes
B002,Sapiens,Yuval Noah Harari,2011,Non-fiction,4.4,no
B003,Atomic Habits,James Clear,2018,Self-help,4.7,yes
B004,The God of Small Things,Arundhati Roy,1997,Fiction,4.3,yes
B005,Thinking Fast and Slow,Daniel Kahneman,2011,Psychology,4.1,no
B006,The Immortals of Meluha,Amish Tripathi,2010,Mythology,4.0,yes
B007,Deep Work,Cal Newport,2016,Self-help,4.5,yes
B008,A Brief History of Time,Stephen Hawking,1988,Science,4.3,no
B009,The White Tiger,Aravind Adiga,2008,Fiction,4.1,yes
B010,Outliers,Malcolm Gladwell,2008,Non-fiction,4.2,yes
"""

# Write to disk
with open("library.csv", "w") as lib_file:
    lib_file.write(csv_content)

print("library.csv created.")
print()

# Let's see the raw file content — this is exactly what's stored on disk
with open("library.csv", "r") as f:
    raw = f.read()

print("--- Raw file content ---")
print(raw)
````

````python
# Method 1: csv.reader — each row comes back as a list
# The first row is the header

print("Reading with csv.reader:")
print("=" * 50)

with open("library.csv", "r") as lib_file:
    reader = csv.reader(lib_file)
    # print(reader)
    for row in reader:
        # print(row)   # each row is a plain Python list
        print(row)
````

### Observations from `csv.reader`

- The first row printed is the **header row** — `['book_id', 'title', 'author', ...]`. It is just another row to `csv.reader`. You have to handle it yourself.
- Every value is a **string**, even the year and rating. Python does not know these should be numbers.
- To access a value, you use an **index**: `row[0]` is `book_id`, `row[2]` is `author`, etc. This is fragile — if column order ever changes, your code breaks.

````python
# Handling headers manually with csv.reader
# Use next() to consume the first row before the loop

print("Reading with csv.reader — header separated:")
print("=" * 50)

with open("library.csv", "r") as f:
    reader = csv.reader(f)

    header = next(reader)     # pulls the first row out and stores it
    print("Header:", header)
    print()

    for row in reader:        # now the loop starts from the second row
        book_id = row[0]
        title   = row[1]
        author  = row[2]
        print(f"  {book_id} | {title:<30} | {author}")
````

````python
# Method 2: csv.DictReader — each row comes back as a dictionary
# Column names become the keys automatically — no manual header handling needed

print("Reading with csv.DictReader:")
print("=" * 50)

with open("library.csv", "r") as f:
    reader = csv.DictReader(f)

    for row in reader:
        print(dict(row))   # each row is a Python dictionary
````

````python
# DictReader in practice: much more readable and robust
# Access values by name, not by index

print(f"{'ID':<6} {'Title':<30} {'Author':<22} {'Year':<6} {'Genre':<14} {'Rating':<8} {'Available'}")
print("-" * 100)

with open("library.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(f"{row['book_id']:<6} {row['title']:<30} {row['author']:<22} {row['year']:<6} {row['genre']:<14} {row['rating']:<8} {row['available']}")
````

### Whiteboard: `csv.reader` vs `csv.DictReader`

| | `csv.reader` | `csv.DictReader` |
|---|---|---|
| Each row is | A list | A dictionary |
| Access values by | Index (`row[2]`) | Name (`row['author']`) |
| Handles header | Manually with `next()` | Automatically |
| Readable code | Harder | Easier |
| When column order changes | Code breaks | Code still works |

In practice: **use `DictReader` almost always.** It is more readable, safer, and does not require you to memorise column positions.

`csv.reader` is useful when you are dealing with very simple files, performance-critical code, or files with no headers.

## A.2 — Writing a CSV File

Writing CSVs is the reverse of reading. Instead of `csv.reader` / `csv.DictReader`, we use `csv.writer` / `csv.DictWriter`.

The same logic applies:
- `csv.writer` — you provide each row as a list
- `csv.DictWriter` — you provide each row as a dictionary, and it maps keys to columns for you

One important detail when opening a file for writing:
- Use `open(filename, "w", newline="")` — the `newline=""` prevents Python from adding extra blank lines on Windows

````python
# Scenario: the library just acquired 3 new books
# We need to write them to a new CSV file

new_books = [
    {"book_id": "B011", "title": "Zero to One",         "author": "Peter Thiel",       "year": "2014", "genre": "Business",    "rating": "4.3", "available": "yes"},
    {"book_id": "B012", "title": "The Lean Startup",    "author": "Eric Ries",          "year": "2011", "genre": "Business",    "rating": "4.1", "available": "yes"},
    {"book_id": "B013", "title": "Ikigai",              "author": "Hector Garcia",      "year": "2016", "genre": "Self-help",   "rating": "4.2", "available": "no"},
]

# Define the column names — this is the header row
fieldnames = ["book_id", "title", "author", "year", "genre", "rating", "available"]

with open("new_arrivals.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)

    writer.writeheader()          # writes the header row first
    writer.writerows(new_books)   # writes all rows at once

print("new_arrivals.csv written.")
print()

# Verify by reading it back
with open("new_arrivals.csv", "r") as f:
    print(f.read())
````

````python
# Real scenario: merging the new arrivals into the main library catalogue
# Open in append mode ('a') to add rows without overwriting existing content

fieldnames = ["book_id", "title", "author", "year", "genre", "rating", "available"]

with open("library.csv", "a", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    # Note: no writeheader() here — we do NOT want to re-write the header
    writer.writerows(new_books)

# Verify: count the total rows now
with open("library.csv", "r") as f:
    reader = csv.DictReader(f)
    all_books = list(reader)

print(f"Total books in library.csv: {len(all_books)}")
print("Last 3 entries:")
for book in all_books[-3:]:
    print(f"  {book['book_id']} — {book['title']} by {book['author']}")
````

````python
# Another real scenario: export only the available books to a separate file
# This is a very common pattern — read, filter, write

fieldnames = ["book_id", "title", "author", "genre", "rating"]

with open("library.csv", "r") as infile, open("available_books.csv", "w", newline="") as outfile:

    reader = csv.DictReader(infile)
    writer = csv.DictWriter(outfile, fieldnames=fieldnames, extrasaction="ignore")
    # extrasaction='ignore' tells DictWriter to skip fields
    # that are in the row but not in our fieldnames list

    writer.writeheader()

    available_count = 0
    for row in reader:
        if row["available"] == "yes":
            writer.writerow(row)
            available_count += 1

print(f"Wrote {available_count} available books to available_books.csv")
print()

with open("available_books.csv", "r") as f:
    print(f.read())
````

## A.3 — Where CSV Falls Short

CSV is excellent for flat, tabular data — but it has real limitations:

**Problem 1: It cannot represent nested data.**

What if each book had multiple authors? Or a list of tags? You cannot put a list inside a CSV cell in any clean way.

**Problem 2: There are no data types.**

Everything is a string. `"4.2"` and `"yes"` and `"1988"` are all just text. You always have to convert manually.

**Problem 3: No standard for complex structures.**

What if a book record needed to include the publisher's address, each chapter's title, and user reviews? CSV cannot express this cleanly.

This is exactly where **JSON** comes in. Let's look at the same library data in JSON format and see how the representation differs.

# PART B — JSON: Working with Semi-Structured Data

## What is semi-structured data?

**Semi-structured** data does not require every record to have exactly the same fields. It has some organisation (keys, tags, nesting), but it is flexible — different records can have different shapes.

Compare:
- **Structured (CSV):** Every book row must have exactly 7 columns, in the same order, of the same type.
- **Semi-structured (JSON):** One book record can have 7 fields. Another can have 9. One book can have a list of tags; another might have no tags at all.

---

## What is JSON?

**JSON** stands for **JavaScript Object Notation**. Despite the name, it has nothing to do with JavaScript in practice — it is the universal format for data exchange on the web.

If you have ever:
- Used a weather app
- Made a payment through UPI
- Loaded a Google Maps result
- Used a chatbot

...then JSON was almost certainly the format in which data was being passed around behind the scenes.

---

## JSON syntax in 5 minutes

JSON has only 6 building blocks. If you know these 6, you know JSON.

| Type | JSON syntax | Python equivalent |
|---|---|---|
| String | `"hello"` | `str` |
| Number | `42` or `3.14` | `int` or `float` |
| Boolean | `true` or `false` | `True` or `False` |
| Null | `null` | `None` |
| Array | `[1, 2, 3]` | `list` |
| Object | `{"key": "value"}` | `dict` |

Two things to watch:
- JSON **strings always use double quotes** — `"hello"` is valid, `'hello'` is not
- JSON booleans are lowercase: `true` / `false`, not `True` / `False`

````python
# Let's see all 6 JSON types in one example
# This is a raw JSON string (just text — Python does not know its structure yet)

json_string = '''
{
    "title":     "Atomic Habits",
    "year":      2018,
    "rating":    4.7,
    "available": true,
    "film":      null,
    "author": {
        "name": "something",
        "age": 30
    },
    "tags":      ["productivity", "psychology", "self-improvement"]
}
'''

# json.loads() — "load string" — parses JSON text into a Python object
book = json.loads(json_string)

print("Type after parsing:", type(book))
print()
for key, value in book.items():
    print(f"  {key:<12} = {str(value):<30}  (Python type: {type(value).__name__})")
````

### Notice the type mapping

When `json.loads()` parses JSON:
- JSON `true` becomes Python `True`
- JSON `null` becomes Python `None`
- JSON numbers become Python `int` or `float` automatically — **no manual conversion needed**
- JSON arrays become Python `list`
- JSON objects become Python `dict`

This is a big advantage over CSV, where everything was a string.

## B.1 — Nested Data

The most important feature JSON has that CSV does not: **nesting**.

In JSON, a value can itself be an object or an array. This lets you represent complex real-world data naturally.

Here is the same `Atomic Habits` record with richer nested data:

````python
# A richer book record with nested structure

nested_json = '''
{
    "book_id": "B003",
    "title":   "Atomic Habits",
    "author": {
        "name":      "James Clear",
        "nationality": "American",
        "website":   "jamesclear.com"
    },
    "publication": {
        "year":      2018,
        "publisher": "Avery",
        "pages":     320
    },
    "genres": ["Self-help", "Psychology", "Productivity"],
    "ratings": {
        "average":  4.7,
        "count":    98234
    },
    "reviews": [
        {"user": "Ananya", "score": 5, "comment": "Changed my daily routine completely."},
        {"user": "Rohan",  "score": 4, "comment": "Very practical and easy to apply."}
    ],
    "available": true
}
'''

book = json.loads(nested_json)

# Accessing flat fields — same as before
print("Title:    ", book["title"])
print("Available:", book["available"])
print()

# Accessing a nested object — use two keys
print("Author name:       ", book["author"]["name"])
print("Author nationality:", book["author"]["nationality"])
print("Publisher:         ", book["publication"]["publisher"])
print("Pages:             ", book["publication"]["pages"])
print()

# Accessing a list field
print("Genres:", book["genres"])
print("First genre:", book["genres"][0])
print()

# Accessing a list of objects — combine list index with dict key
print("Reviews:")
for review in book["reviews"]:
    print(f"  {review['user']} gave {review['score']}/5 — '{review['comment']}'")
````

````python
# Mental model: think of nested JSON as a set of boxes inside boxes
# Each key opens the next box. Lists are numbered boxes.

# Practice: what does each line print?
print(book["ratings"]["average"])             # average rating
print(book["reviews"][1]["user"])             # second reviewer's name
print(len(book["genres"]))                    # number of genres
print(book["publication"]["year"])            # publication year
print(book["reviews"][0]["score"])            # first reviewer's score
````

## B.2 — Parsing JSON Strings

### The two core functions

Python's `json` module has two pairs of functions. The naming follows a simple pattern:

| Function | Input | Output | Memory aid |
|---|---|---|---|
| `json.loads(s)` | JSON **string** | Python object | **loads** = load from string |
| `json.load(f)` | File **handle** | Python object | **load** = load from file |
| `json.dumps(obj)` | Python object | JSON **string** | **dumps** = dump to string |
| `json.dump(obj, f)` | Python object + file | Writes to file | **dump** = dump to file |

The `s` in `loads` and `dumps` stands for **string**. That is the only difference.

````python
# json.loads() — parsing a JSON string that arrived from somewhere
# (e.g. from an API response, from a database, from a config value)

# Imagine this string came from an API
api_response_text = '{"book_id": "B001", "title": "The Alchemist", "rating": 4.2, "available": true}'

print("Type before parsing:", type(api_response_text))
print("Value before parsing:", api_response_text)
print()

parsed = json.loads(api_response_text)

print("Type after parsing:", type(parsed))
print("Value after parsing:", parsed)
print()

# Now we can work with it like a normal Python dictionary
print("Title:",   parsed["title"])
print("Rating:",  parsed["rating"],  " -> type:", type(parsed["rating"]).__name__)
print("Available:", parsed["available"], " -> type:", type(parsed["available"]).__name__)
````

````python
# json.load() — reading JSON from a file
# First, let's create a JSON file to read

# Our full library as a list of simple records (flat — same fields as the CSV)
library_list = [
    {"book_id": "B001", "title": "The Alchemist",           "author": "Paulo Coelho",        "year": 1988, "genre": "Fiction",     "rating": 4.2, "available": True},
    {"book_id": "B002", "title": "Sapiens",                 "author": "Yuval Noah Harari",   "year": 2011, "genre": "Non-fiction", "rating": 4.4, "available": False},
    {"book_id": "B003", "title": "Atomic Habits",           "author": "James Clear",          "year": 2018, "genre": "Self-help",   "rating": 4.7, "available": True},
    {"book_id": "B004", "title": "The God of Small Things", "author": "Arundhati Roy",        "year": 1997, "genre": "Fiction",     "rating": 4.3, "available": True},
    {"book_id": "B005", "title": "Thinking Fast and Slow",  "author": "Daniel Kahneman",      "year": 2011, "genre": "Psychology",  "rating": 4.1, "available": False},
]

# Write to a JSON file first
with open("library.json", "w") as f:
    json.dump(library_list, f, indent=2)

print("library.json written. Raw content:")
print()
with open("library.json", "r") as f:
    print(f.read())
````

````python
# Now read it back with json.load()

with open("library.json", "r") as f:
    books = json.load(f)

print(f"Loaded {len(books)} books")
print(f"Type: {type(books)}")
print()

# Key difference from CSV: types are preserved
first = books[0]
print("First book:")
for key, value in first.items():
    print(f"  {key:<12} = {str(value):<25}  type: {type(value).__name__}")
````

| `rating` | `"4.2"` (string) | `4.2` (float) |

With CSV you always had to write `float(row['rating'])` before doing any math. With JSON, the type is already correct when you load it.

## B.3 — Converting Python Dicts to JSON

So far we have gone **JSON → Python** (parsing). Now let's go **Python → JSON** (serialising).

This is what you do when you want to:
- Save Python data to a file in a portable format
- Send data to a web API
- Write a configuration file
- Share data with another program

The function is `json.dumps()` (to a string) or `json.dump()` (to a file).

````python
# json.dumps() — convert a Python object to a JSON string

book_record = {
    "book_id":   "B014",
    "title":     "The Psychology of Money",
    "author":    "Morgan Housel",
    "year":      2020,
    "rating":    4.6,
    "available": True,
    "tags":      ["finance", "behavioural", "investing"]
}

# Basic conversion
json_string = json.dumps(book_record)
print("Basic output:")
print(json_string)
print()

# With indentation — human-readable formatting
pretty_string = json.dumps(book_record, indent=2)
print("Pretty output (indent=2):")
print(pretty_string)
````

## B.4 — The Same Data in Two Formats

Let's make the connection between CSV and JSON explicit. We have been working with the same library data throughout. Now let's look at one record side-by-side in both formats, then write a function that converts our CSV into a proper JSON file.

### One record — two representations

**As a CSV row:**
```
B003,Atomic Habits,James Clear,2018,Self-help,4.7,yes
```

**As a JSON object:**
```json
{
  "book_id":   "B003",
  "title":     "Atomic Habits",
  "author":    "James Clear",
  "year":      2018,
  "genre":     "Self-help",
  "rating":    4.7,
  "available": true
}
```

The content is identical. But notice what changed:
- `year` is now an integer, not a string
- `rating` is now a float, not a string
- `available` is now a boolean `true`, not the string `"yes"`
- Each field has its name right next to its value — no need to remember column order

````python
# Convert library.csv to library_converted.json
# While converting, fix the data types properly

def csv_to_json(csv_filename, json_filename):
    """
    Reads a CSV library file and writes a JSON file with correct types.
    Converts: year -> int, rating -> float, available -> bool
    """
    books = []

    with open(csv_filename, "r") as f:
        reader = csv.DictReader(f)
        for row in reader:
            books.append({
                "book_id":   row["book_id"],
                "title":     row["title"],
                "author":    row["author"],
                "year":      int(row["year"]),          # string -> integer
                "genre":     row["genre"],
                "rating":    float(row["rating"]),      # string -> float
                "available": row["available"] == "yes" # string -> boolean
            })

    with open(json_filename, "w") as f:
        json.dump(books, f, indent=2)

    return len(books)


count = csv_to_json("library.csv", "library_converted.json")
print(f"Converted {count} records from CSV to JSON.")
print()

# Verify: show the first record and confirm types
with open("library_converted.json", "r") as f:
    converted = json.load(f)

first = converted[0]
print("First record in the JSON output:")
for key, value in first.items():
    print(f"  {key:<12} = {str(value):<25}  type: {type(value).__name__}")
````

````python
# And the reverse: JSON back to CSV
# This is also a common real-world task

def json_to_csv(json_filename, csv_filename):
    """
    Reads a JSON library file and writes it back as CSV.
    """
    with open(json_filename, "r") as f:
        books = json.load(f)

    fieldnames = ["book_id", "title", "author", "year", "genre", "rating", "available"]

    with open(csv_filename, "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        for book in books:
            # Convert bool back to yes/no for CSV
            row = dict(book)
            row["available"] = "yes" if book["available"] else "no"
            writer.writerow(row)

    return len(books)


count = json_to_csv("library_converted.json", "library_roundtrip.csv")
print(f"Converted {count} records from JSON back to CSV.")
print()
print("First 3 rows of the roundtrip CSV:")
with open("library_roundtrip.csv", "r") as f:
    reader = csv.DictReader(f)
    for i, row in enumerate(reader):
        if i >= 3:
            break
        print(f"  {dict(row)}")
````

## Session Summary

### Part A — CSV

| Concept | Key point |
|---|---|
| What CSV is | Plain text file; commas separate values; first row is the header |
| `csv.reader` | Rows as lists; access by index; use `next()` for the header |
| `csv.DictReader` | Rows as dicts; access by name; header handled automatically — prefer this |
| `csv.DictWriter` | Write rows from dicts; call `writeheader()` once; use append mode `"a"` to add rows |
| Types in CSV | Everything is a string — always convert manually (`int()`, `float()`) |

---

### Part B — JSON

| Concept | Key point |
|---|---|
| What JSON is | Text format for semi-structured data; supports nesting; universal web format |
| 6 types | string, number, boolean, null, array, object |
| Nesting | Values can be objects or arrays; access with chained keys/indexes |
| `json.loads(s)` | JSON string → Python object |
| `json.load(f)` | JSON file → Python object |
| `json.dumps(obj)` | Python object → JSON string (`indent=2` for readable output) |
| `json.dump(obj, f)` | Python object → JSON file |
| Types in JSON | Preserved on load — no manual conversion needed |

---

### When to use which format

| Situation | Use |
|---|---|
| Flat tabular data (rows and columns) | CSV |
| Opening in Excel or Google Sheets | CSV |
| Nested or variable-structure data | JSON |
| Data from a web API | JSON |
| Configuration files | JSON |
| Passing data between two programs | JSON |
| Maximum human readability for simple data | CSV |
