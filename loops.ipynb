```python
fruits = ("apple", "banana", "mango") #prints index and values from lists works for lists and tuples
for i in range(len(fruits)):
  print(i,fruits[i])

for j,fruit in enumerate(fruits): #preferred over the above same as above
  print(j,fruit)
```

```python
vegetables=["carrot","cucumber","broccoli"]
for index,value in enumerate(vegetables,start=1): #start with index 1
  print(index,value)
```

```python
student = {"name": "Shaan", "age": 21,"dob":"17-09-2004"}
for keys in student: # prints keys
  print(keys)
for val in student.values(): #prints values of key
  print(val)
for key,value in student.items():
  print(f"{key}:{value}") # prints both
print(student)
res=student.get("name")
print(res)
```
