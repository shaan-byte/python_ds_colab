Reading CSV file into dataframes

```python
# Generate and save a realistic CSV file
# Students would normally receive this file from a data source or instructor
# We create it here so the notebook is self-contained

import csv, random
random.seed(42)

departments = ["Engineering", "Sales", "HR", "Finance"]
cities      = ["Mumbai", "Delhi", "Bangalore", "Chennai", "Hyderabad"]
first_names = ["Ananya", "Rohit", "Priya", "Vikram", "Sneha", "Arjun",
               "Deepa", "Kiran", "Meera", "Suresh", "Lakshmi", "Rahul",
               "Divya", "Arun", "Pooja", "Sanjay", "Nisha", "Tarun"]
last_names  = ["Sharma", "Mehta", "Nair", "Rao", "Pillai", "Das",
               "Joshi", "Iyer", "Bose", "Kumar", "Reddy", "Singh"]

dept_salary_ranges = {
    "Engineering": (70000, 150000),
    "Sales":       (55000, 110000),
    "HR":          (50000,  95000),
    "Finance":     (65000, 130000),
}

rows = []
for i in range(50):
    emp_id  = 1001 + i
    name    = random.choice(first_names) + " " + random.choice(last_names)
    dept    = random.choice(departments)
    low, hi = dept_salary_ranges[dept]
    salary  = round(random.uniform(low, hi), -3)  # rounded to nearest 1000
    exp     = random.randint(1, 20)
    city    = random.choice(cities)
    perf    = round(random.uniform(2.0, 5.0), 1)  # performance rating
    year    = random.randint(2004, 2023)
    month   = random.randint(1, 12)
    join    = f"{year}-{month:02d}-01"
    active  = random.choice([True, True, True, False])  # 75% active

    # Introduce realistic messiness in ~15% of rows
    if random.random() < 0.08:
        salary = ""           # missing salary
    if random.random() < 0.08:
        perf = ""             # missing performance rating
    if random.random() < 0.06:
        dept = dept.lower()   # casing inconsistency

    rows.append([emp_id, name, dept, salary, exp, city, perf, join, active])

with open("employees.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["employee_id", "name", "department", "salary",
                     "experience", "city", "performance_rating", "join_date", "active"])
    writer.writerows(rows)

print("employees.csv created — 50 rows, 9 columns.")
print("Let's now load it with Pandas.")
```

Summarizing and Grouping data

```python
random.seed(42)
departments = ["Engineering", "Sales", "HR", "Finance"]
cities      = ["Mumbai", "Delhi", "Bangalore", "Chennai", "Hyderabad"]
first_names = ["Ananya","Rohit","Priya","Vikram","Sneha","Arjun",
               "Deepa","Kiran","Meera","Suresh","Lakshmi","Rahul",
               "Divya","Arun","Pooja","Sanjay","Nisha","Tarun"]
last_names  = ["Sharma","Mehta","Nair","Rao","Pillai","Das",
               "Joshi","Iyer","Bose","Kumar","Reddy","Singh"]
dept_salary = {"Engineering":(70000,150000),"Sales":(55000,110000),
               "HR":(50000,95000),"Finance":(65000,130000)}
rows = []
for i in range(50):
    dept   = random.choice(departments)
    lo, hi = dept_salary[dept]
    salary = round(random.uniform(lo, hi), -3)
    perf   = round(random.uniform(2.0, 5.0), 1)
    year   = random.randint(2004, 2023)
    month  = random.randint(1, 12)
    row    = [1001+i,
              random.choice(first_names)+" "+random.choice(last_names),
              dept if random.random()>=0.06 else dept.lower(),
              "" if random.random()<0.08 else salary,
              random.randint(1,20),
              random.choice(cities),
              "" if random.random()<0.08 else perf,
              f"{year}-{month:02d}-01",
              random.choice([True,True,True,False])]
    rows.append(row)
with open("employees.csv", "w", newline="") as f:
    w = csv.writer(f)
    w.writerow(["employee_id","name","department","salary",
                "experience","city","performance_rating","join_date","active"])
    w.writerows(rows)

emp = pd.read_csv("employees.csv", parse_dates=["join_date"], index_col="employee_id")
emp["salary"]             = pd.to_numeric(emp["salary"],             errors="coerce")
emp["performance_rating"] = pd.to_numeric(emp["performance_rating"], errors="coerce")
emp["department"]         = emp["department"].str.title()

# Add a seniority column — used throughout today's session
emp["seniority"] = pd.cut(
    emp["experience"],
    bins   = [0, 3, 8, 100],
    labels = ["Junior", "Mid", "Senior"]
)

print("Dataset loaded.")
print(f"Shape: {emp.shape}")
print(emp.head(4))
```

Methods

```python
# Individual methods on a single column
sal = emp["salary"]

print("Salary statistics:")
print(f"  count:   {sal.count()}        (non-missing values)")
print(f"  mean:    {sal.mean():,.0f}")
print(f"  median:  {sal.median():,.0f}")
print(f"  std:     {sal.std():,.0f}")
print(f"  min:     {sal.min():,.0f}")
print(f"  max:     {sal.max():,.0f}")
print(f"  range:   {sal.max() - sal.min():,.0f}")
print()

# Percentiles — .quantile() gives you any percentile you ask for
print("Percentiles:")
print(f"  25th percentile: {sal.quantile(0.25):,.0f}")
print(f"  50th percentile: {sal.quantile(0.50):,.0f}   (same as median)")
print(f"  75th percentile: {sal.quantile(0.75):,.0f}")
print(f"  90th percentile: {sal.quantile(0.90):,.0f}")
```

Mean and Median

```python
# Mean vs median — why the difference matters
# If the mean and median are far apart, there are likely outliers pulling the mean

mean_sal   = emp["salary"].mean()
median_sal = emp["salary"].median()

print(f"Mean salary:   {mean_sal:,.0f}")
print(f"Median salary: {median_sal:,.0f}")
print(f"Difference:    {abs(mean_sal - median_sal):,.0f}")
print()

if abs(mean_sal - median_sal) > 5000:
    print("The mean and median are notably different.")
    print("This suggests a skewed salary distribution — some very high or low values")
    print("are pulling the mean away from the typical employee's salary.")
    print("In such cases, the median is often a more honest 'typical' value.")
else:
    print("Mean and median are close — the salary distribution is fairly symmetric.")
```

Correlation

```python
#corr checks how closely two columns are related to one another values rnage from -1 to +1
correlation=emp['salary'].corr(emp['experience'])
print(correlation) #closer to -1 inversely proportional closer to +1 directly proportional
```

```python
# .corr() — correlation between numeric columns
# Values range from -1 (perfect negative) to +1 (perfect positive)
# 0 means no linear relationship

print("Correlation matrix (numeric columns):")
print(emp.select_dtypes(include=["number"]).corr().round(2))
print()
print("Reading this: check if salary and experience are correlated.")
sal_exp_corr = emp["salary"].corr(emp["experience"])
print(f"salary vs experience correlation: {sal_exp_corr:.3f}")
if sal_exp_corr > 0.3:
    print("Positive correlation — more experienced employees tend to earn more.")
else:
    print("Weak correlation — experience alone does not strongly predict salary here.")
```

```python
gbd=emp.groupby('department')['salary'].mean()
print(f"the mean salary for each department \n {gbd}") #lets you groupby data based on the argument to perform op

emp_count=emp.groupby('department')['salary'].count()
print(f"the number of employees in each department \n {emp_count}")
```
