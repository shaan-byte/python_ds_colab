```python
set1={1,2,3,4,5}
print(type(set1))
s=set()
s.add(101) #adding elements in set
s.add(102)
set2={2,3,6,7}
print(set1 & set2) #intersection set1.intersection(set2)
print(set1 | set2) #union set1.union(set2)
print(set1 - set2) #differnece elemetns in set1 an not in set2
print(set1 ^ set2) #symmetric differnece elemetns that are exactly in one of the sets
num={40,30,20,10,2222}
num.update([69,34,221])
num.discard(40) #discard or remove remove gives error when element not present
#disjoint
odd_A={1,3,5,7,9}
even_B={2,4,6,8,10}
print(odd_A.isdisjoint(even_B))
print(num)

#superset/subset
small_set={1,2}
large_set={1,2,3,4}
print(small_set.issubset(large_set))
print(large_set.issuperset(small_set))
```
