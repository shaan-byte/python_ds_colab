```python
import csv
import json
import os
from io import StringIO
```

```python
# First, let's create our CSV file on disk so we have something to read

csv_content = """book_id,title,author,year,genre,rating,available
B001,The Alchemist,Paulo Coelho,1988,Fiction,4.2,yes
B002,Sapiens,Yuval Noah Harari,2011,Non-fiction,4.4,no
B003,Atomic Habits,James Clear,2018,Self-help,4.7,yes
B004,The God of Small Things,Arundhati Roy,1997,Fiction,4.3,yes
B005,Thinking Fast and Slow,Daniel Kahneman,2011,Psychology,4.1,no
B006,The Immortals of Meluha,Amish Tripathi,2010,Mythology,4.0,yes
B007,Deep Work,Cal Newport,2016,Self-help,4.5,yes
B008,A Brief History of Time,Stephen Hawking,1988,Science,4.3,no
B009,The White Tiger,Aravind Adiga,2008,Fiction,4.1,yes
B010,Outliers,Malcolm Gladwell,2008,Non-fiction,4.2,yes
"""

with open("library.csv","w") as f:
  f.write(csv_content)

with open("library.csv","r") as f:
  raw=f.read()

print(raw)
```

```python
#csv reader returns csv file as list

with open("library.csv","r") as lib_file:
  reader=csv.reader(lib_file)
  for row in reader:
    #prints as list
    print(row)
```

```python
#csv write for lists

rows = [
    ["Book", "Author", "Price"],
    ["Python", "Guido", 500],
    ["C++", "Bjarne", 700]
]


with open("practice.csv","w",newline="") as file:
  writer=csv.writer(file)
#as multiple rows
  writer.writerows(rows)
#as row by row
  writer.writerow(["kotlin","Dennis",1200])
```

```python
#csv dictwrite for dictionaries
books=[{
    "book_id": "B001",
    "title": "The Alchemist",
    "author": "Paulo Coelho",
    "year": 1988,},{"book_id":"B002","title":"wimpy kid","author":"James Franco","year":2000}
]

with open("books.csv","w",newline="") as file:
  #write fieldnames
  fieldnames=["book_id","title","author","year"]
  #dictwriter for dictionaries
  writer1=csv.DictWriter(file,fieldnames=fieldnames)
  #write header
  writer1.writeheader()
  #write rows
  writer1.writerows(books)

```

```python
# Let's see all 6 JSON types in one example
# This is a raw JSON string (just text — Python does not know its structure yet)

json_string = '''
{
    "title":     "Atomic Habits",
    "year":      2018,
    "rating":    4.7,
    "available": true,
    "film":      null,
    "author": {
        "name": "something",
        "age": 30
    },
    "tags":      ["productivity", "psychology", "self-improvement"]
}
'''

# json.loads() — "load string" — parses JSON text into a Python object
book = json.loads(json_string)
print(book)
print(type(book))
#json dumps python object to json string
#json.dump(obj,f) takes object and file and writes to a file
```
